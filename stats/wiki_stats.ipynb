```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import numpy

# Apply the default theme at seaborn
sns.set_theme()
```

```python
# Opening JSON file
file_name = './../output/output_wiki.json'

df = pd.read_json(file_name)
```

```python
df
```

```python
df = df.applymap(lambda x: None if len(x)==0 else x)
```

```python
df
```

```python
df.isnull().sum()
```

```python
print("Number of cells:", df.size)
print("Number of value None:", df.isnull().sum().sum())
print("Number of row:", df.shape[0])
print("Number of columns:", df.shape[1])
```

```python
# to extract the first elements from list
df.type = df.type.apply(lambda x: x[0] if x is not None else x)
```

```python
#type
print(df['type'].value_counts())
type_companies = df['type'].value_counts()
type(type_companies)
#the top 10 types of companies are taken here because there are many types
type_companies_top10 = type_companies.head(10)
keys = type_companies_top10.index
values = type_companies_top10.values

print(keys)
print(values)

#graph with mathplotlib
fig, ax = plt.subplots()
width = 0.75
fig.set_size_inches(10, 5)
ind = np.arange(len(values))
ax.barh(keys, values, width, color = "green")

for i, v in enumerate(values):
    ax.text(v + 1, i - 0.2, str(v),
            color = 'green', fontweight = 'bold')
plt.ylabel("top 10 type of companies")
plt.xlabel("number of companies")
plt.show()
```

```python
# to extract the first elements from list
df.industry = df.industry.apply(lambda x: x[0] if x is not None else x)
```

```python
#industry
print(df['industry'].value_counts())
industry_companies = df['industry'].value_counts()
type(industry_companies)
#the top 10 industry
industry_companies_top10 = industry_companies.head(10)
keys = industry_companies_top10.index
values = industry_companies_top10.values

print(keys)
print(values)

#graph with mathplotlib
fig, ax = plt.subplots()
width = 0.75
fig.set_size_inches(10, 5)
ind = np.arange(len(values))
ax.barh(keys, values, width, color = "green")

for i, v in enumerate(values):
    ax.text(v + 1, i - 0.2, str(v),
            color = 'green', fontweight = 'bold')
plt.ylabel("top 10 industry of companies")
plt.xlabel("number of companies")
plt.show()
```

```python
# to extract the first elements from list
df.founded = df.founded.apply(lambda x: x[0] if x is not None else x)

# df.founded.apply(lambda txt: int(s) for s in txt for w in s.split() if w.isdigit() and len(s)==4)
```

```python
companies_to_date = {'<1950': 0, '1950-2000': 0, '>2000': 0}
#generate dictionary with key an interval time and with value number of companies in that time interval
for s in df.founded.values:
    if s is not None:
        for w in s.split():
            if s.isdigit() and len(s)==4:
                s = int(s)
                type(s)
                if s<1950:
                    companies_to_date['<1950'] += 1
                elif 1950 < s < 2000:
                    companies_to_date['1950-2000'] += 1
                else:
                    companies_to_date['>2000'] += 1

print(companies_to_date)
companies_to_date = pd.Series(companies_to_date)
keys = companies_to_date.index
values = companies_to_date.values
print(keys)
print(values)

#graph with mathplotlib
fig, ax = plt.subplots()
width = 0.75
fig.set_size_inches(10, 5)
ind = np.arange(len(values))
ax.barh(keys, values, width, color = "green")

for i, v in enumerate(values):
    ax.text(v + 1, i - 0.05, str(v),
            color = 'green', fontweight = 'bold')
plt.ylabel("fondation date")
plt.xlabel("number of companies")
plt.show()
```
